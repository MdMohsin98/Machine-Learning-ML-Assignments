# Web Secrapping

```python
# Importing Required Libraries
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import pandas as pd
```

```python
# Reading the file
df = pd.read_csv('Input.csv', index_col='URL_ID')
```

```python
# Showing the file
df.head()
```

```python
# Reading the website for scrapping
url1 = df.URL[1]
```

```python
# Accessing the url with requests object
result = requests.get(url1)
# Scrape the webpage document
webpage1 = result.content
# Creating a soup object for parsing web page using html parser
sl_soup = BeautifulSoup(webpage1, 'html.parser')
```

```python
# Closing the result object
result.close()
```

```python
# Viewing the content of the soup object
sl_soup.contents
```

```python
# Prettify the contents
print(sl_soup.prettify())
```
